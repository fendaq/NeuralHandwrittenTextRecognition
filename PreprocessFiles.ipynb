```python
from glob import glob
import os

extension = 'png'
root = './data/IAM'
structure = 'words_original'
path = os.path.join(root, 'imgs', structure) 

files = []
for root, dirs, flst in os.walk(path):
    for file in flst:
        if file.endswith(extension):
            files.append(os.path.join(root, file))
                
files = sorted(files)               

len(files)
```

```python
from multiprocessing import Pool
import multiprocessing 

MAXCPU = multiprocessing.cpu_count()

def task(file):
    out_dir = './data/IAM/imgs/words/'
    name = os.path.basename(file).split('.')[0]

    command = "imgtxtenh -d 118.110 {0} /tmp/{1}.png".format(file, name)
    os.system(command)
    
    command = ("convert /tmp/{0}.png -deskew 40% \
    -bordercolor white -border 5 -trim -bordercolor white -border 20x0 \
    +repage -strip {1}{0}.jpg").format(name, out_dir)
    errno = os.system(command)

    return 0
```

```python
pool = Pool(MAXCPU)

res = pool.map(task, files, chunksize = 1)
```
